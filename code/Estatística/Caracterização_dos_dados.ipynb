## BoxPlot

#### Dados de vazão, boxplot com outilers;

```python
dados.boxplot(column=["MANSO", "MANSO_JUS", "COIMBRA_F"], vert=False, grid=False, figsize=(12, 4))
plt.show()
```

#### Dados de vazão, boxplot sem outilers;

```python
dados.boxplot(column=["MANSO", "MANSO_JUS", "COIMBRA_F"], vert=False, showfliers=False, grid=False)
plt.show()
```

#### Dados de precipitação, boxplot com outilers;

```python
dados.boxplot(column=["COIMBRA_P"], vert=False, showfliers=True, grid=False)
plt.show()
```

#### Dados de vazão, boxplot sem outilers;

```python
dados.boxplot(column=["COIMBRA_P"], vert=False, showfliers=False, grid=False)
plt.show()
```
